```python
def inactivation_bars(trace):
    trace = np.abs(trace)
    timepoints = np.array([250, 1000, 5000, 10000])
    timepoints *= 10
    peak_current = np.max(trace)
    res = []
    for time in timepoints:
        res.append(100 - 100 * trace[time]/peak_current)
    res = np.array(res)
    return res
```

```python
#  https://stackoverflow.com/a/37664693/13213091
def wrapped_ndptr(*args, **kwargs):
    base = np.ctypeslib.ndpointer(*args, **kwargs)
    def from_param(cls, obj):
        if obj is None:
            return obj
        return base.from_param(obj)
    return type(base.__name__, (base,), {'from_param': classmethod(from_param)})
DoubleArrayType_1D = wrapped_ndptr(dtype=np.float64, ndim=1, flags='C_CONTIGUOUS')
DoubleArrayType_2D = wrapped_ndptr(dtype=np.float64, ndim=2, flags='C_CONTIGUOUS')
```

```python
def get_period(data, threshold):
    L = len(data)
    depols = []
    for i in range(L - 1):
        if data[i] <= threshold and data[i + 1] >= threshold:
            depols.append(i)
    T_sum = depols[-1] - depols[0]
    return T_sum / (len(depols) - 1)
```

```python
def single_pulse(S, C, holding_potential = -60.0, pulse_potential = -10.0, pulse_length = 100):
    C["V_factor"] = 0.0
    S[1] = holding_potential
    
    t_sampling = 1
    CL = 1000
    C['CL'] = CL
    stim_period = C['CL']
    n_samples_per_stim = int(stim_period / t_sampling)
    n_beats = 1
    tol = 1e-7
    
    output_HP = np.zeros((n_samples_per_stim * n_beats + 1, len(S)))
    output_HP_A = np.zeros((n_samples_per_stim * n_beats + 1, 31))
    
    status = run(S.astype(float).copy(), C.values.astype(float).copy(), n_beats, t_sampling, tol, output_HP, output_HP_A)
    print(status, end = '')
    
    S = output_HP[-1, :]
    S[1] = pulse_potential
    
    t_sampling = 0.1
    CL = pulse_length
    C['CL'] = CL
    stim_period = C['CL']
    n_samples_per_stim = int(stim_period / t_sampling)
    n_beats = 1
    tol = 1e-7
    
    output_pulse = np.zeros((n_samples_per_stim * n_beats + 1, len(S)))
    output_pulse_A = np.zeros((n_samples_per_stim * n_beats + 1, 31))
    
    status = run(S.astype(float).copy(), C.values.astype(float).copy(), n_beats, t_sampling, tol, output_pulse, output_pulse_A)
    print(status, end = '')
    
    output_HP = pd.DataFrame(output_HP, columns=legend_states.index)
    output_HP_A = pd.DataFrame(output_HP_A, columns = legend_algebraic.index)
    
    output_pulse = pd.DataFrame(output_pulse, columns=legend_states.index)
    output_pulse_A = pd.DataFrame(output_pulse_A, columns = legend_algebraic.index)
    
    C["V_factor"] = 1.0
    
    return (output_HP, output_HP_A, output_pulse, output_pulse_A)
```

```python
def markers(time, AP):
    RMP = np.min(AP[np.nonzero(time > 0)])
    Vpeak = np.max(AP)
    dVdt = np.divide(np.diff(AP), np.diff(time))
    dVdt_max = np.max(dVdt)
    AP_delta = Vpeak - RMP
    AP_normalized = (AP - RMP)/AP_delta
    APD80 = np.max(time[AP_normalized>0.2]) - np.min(time[AP_normalized>0.1])
    APD50 = np.max(time[AP_normalized>0.5]) - np.min(time[AP_normalized>0.5])
    APD20 = np.max(time[AP_normalized>0.8]) - np.min(time[AP_normalized>0.8])
    raw = [RMP, Vpeak, AP_delta, APD20, APD50, APD80, dVdt_max]
    formatted = {"MDP": RMP, "V_peak" : Vpeak,
                 "AP_delta": AP_delta,
                 "APD20": APD20, "APD50": APD50,
                "APD80": APD80, "dVdt_max": dVdt_max}
    return {"raw":raw, "formatted":formatted}
```
